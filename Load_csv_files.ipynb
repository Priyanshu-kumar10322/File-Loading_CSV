# Reading a CSV file 

- `CSV` - Comma-Separated Values 
- `TSV` - Tab-Seperated Values

### Loading a Local CSV File

```python
import pandas as pd

# Loading a local file
df = pd.read_csv('aug_train.csv')
print(df)
```

### Loading a CSV File from a URL

When a CSV file is hosted on a remote server (like a raw GitHub repository) and requires browser headers to bypass server security, you use the requests library alongside io.StringIO to fetch and stream the string content seamlessly into Pandas

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/..." # Replace with your target URL
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}

# Requesting data from the server
req = requests.get(url, headers=headers)
data = StringIO(req.text)

df = pd.read_csv(data)
print(df)

# url = show be change according to the site
```

### The Sep Parameter (Handling Alternative Delimiters)

```python
# Ingesting a Tab-Separated Values (TSV) dataset
df = pd.read_csv('movie_titles_metadata.tsv', sep='\t')
print(df)
```

### The Names Parameter (Assigning Custom Headers)

```python
# Defining manual headers for a dataset without a column row
column_names = ['sno', 'movie_name', 'release_year', 'rating', 'votes', 'genres']
df = pd.read_csv('movie_titles_metadata.tsv', sep='\t', names=column_names)
print(df)
```

### The index_col Parameter (Setting a Key Index Column)

```python
# Converting an existing ID feature into the primary row index
df = pd.read_csv('aug_train.csv', index_col='enrollee_id')
print(df)
```

### The header Parameter

```python
# Indicating that the actual table headers are located at row index 1
df = pd.read_csv('test.csv', header=1)
df.head()
```

### The usecols Parameter (Memory Optimization)

```python
# Selecting and loading only three specific columns out of many
df = pd.read_csv('aug_train.csv', usecols=['enrollee_id', 'gender', 'education_level'])
df.head()
```

### The nrows Parameter (Setting Row Intake Limits)

```python
# Limiting data load to just the first 100 rows
df = pd.read_csv('aug_train.csv', nrows=100)
df
```

### Skipping Corrupt Records (error_bad_lines)

```python
# Isolating and passing over corrupted line errors seamlessly
df = pd.read_csv('BX-Books.csv', sep=';', on_bad_lines='skip')
df.head()
```

### The dtype Parameter (Explicit Typecasting)

```python
# Typecasting the 'target' attribute directly to an integer on load
df = pd.read_csv('aug_train.csv', dtype={'target': int})
df.info()
```

### The na_values Parameter (Custom Missing Value Definitions)

```python
# Flagging 'Male' text inputs to register as missing records (NaN)
df = pd.read_csv('aug_train.csv', na_values=['Male'])
df.head()
```

### The chunksize Parameter (Streaming Huge Files)

```python
# Segmenting a massive dataset into blocks of 5000 rows
dfs = pd.read_csv('aug_train.csv', chunksize=5000)

# Sequentially applying code logic across each chunk iteration
for chunk in dfs:
    print(chunk.shape)
```
